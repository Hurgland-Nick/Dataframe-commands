 ### Convertir une colonne en numérique

- Methode 1 : La méthode Naïve

```python
# Créons une méthode test pour voir
# Faisons le dans la colonne Installe qui a une valeur du type: 1,000,000+ or nous devons avoir 1000000.
# Dans un premier temps nous allons supprimer les virgules(,) ---> on considère que la colonne est un string

# On applique replace pour remplacer les "+" par du vide
test = google_app_sans_na["Installs"].str.replace("+","")

# On applique replace pour remplacer les "," par du vide
test = test.str.replace(",","")

# On converti la colonne en numérique -----> Methode astype
test = test.astype(int)

# Afficher la variable
# On voit bien que la colonne est plus propre et de type numérique dtype: int32
test.head()
```

- Methode 2 : La méthode pro
     - La Méthode apply()

```python
# J'applique 2 fois replace, je viens de bricoler cela !!! ;-) et cela marche ahahah, je gagne une ligne
# Appliquons directement sur la colonne
google_app_sans_na["Installs"] = google_app_sans_na["Installs"].apply(lambda x: (x.replace("+","").replace(",","")))

# On converti la colonne en Integer
google_app_sans_na["Installs"] = google_app_sans_na["Installs"].astype(int)

# On aurait aussi pu faire avec apply
# google_app_sans_na["Installs"] = google_app_sans_na["Installs"].apply(lambda x: int(x))
```

# Grouper un Dataframe sur une ou plusieurs colonnes (groupby)

```python
# Voir ce que contient ma colonne, pour voir les transformations à faire si nécessaire pour la suite
# Nous voyons que le colonne "Price" porte un dollars devant le nombre et on en veut pas et du type Objet aussi
google_app_sans_na["Price"].unique()
```

```python
# Enlevons le dollars en le remplaçant par rien
google_app_sans_na["Price"] = google_app_sans_na["Price"].apply(lambda x: x.replace("$",""))

# On le convertit en float car c'est des nombre à virgules qui seraient bien de garder comme il est
google_app_sans_na["Price"] = google_app_sans_na["Price"].apply(lambda x: float(x))

# Revoyant la colonne "Price"
google_app_sans_na["Price"].unique()
```

```python
# Vérifions la distribution
# Yes
google_app_sans_na.describe()
```

### Utilisons du Sql pour faire la requête

```python
# Import sql fonctionnality for testing cause i'm proficiency in it
import pandasql as pdsql
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())

q = ''' Select count(App) as nb_app,category 
        from google_app_sans_na 
        group by category 
        order by nb_app desc limit 1'''
# Affichons notre requête
df = pysqldf(q)
df
```

### Utilisons du pandas maintenant pour faire la requête

```python
# Agg permet d'aggréger dans ce cas i.e de compter
# A remarquer que le nom des colonnes ont changé et pour l'aggrégat près du nom simple
# On peut ajouter un head(1) pour voir la première colonne.
# <=>  Select count(App) as nb_app,category from google_app_sans_na group by category order by nb_app desc limit 1
google_app_sans_na.groupby("Category").agg(['count']).sort_values(by = ('App', 'count'), ascending = False)
```

```python
# Tentons de l'afficher par un graphe
# Import de matplotlib
%matplotlib inline

#Graphe
google_app_sans_na.groupby("Category").agg(['count']).plot(kind='bar')
```
